```python
#importing necessary packages

import numpy as np
import os
from matplotlib import pyplot as plt
import cv2
import random
import pickle
```

```python
file_list = []
class_list = []
IMG_SIZE = 50 # The size of images the network will work on
```

```python
DATADIR_LIST = [r'..\labels\Labelled Dataset - Fig 51', r'..\labels\Labelled Dataset - Fig 11', r'..\labels\Labelled Dataset - Fig 6']
for DATADIR in DATADIR_LIST:
    for directoryfile in os.listdir(DATADIR):
        print(directoryfile)
```

```python
#All the categories that should be detect in neural network 

CATEGORIES = set()
ignore_files = ['1 - Multipart','2 - Unknown', '2 - Multipart', '3 - Unidentified', '1 - Original Dataset' ] #This files/folder should not include in neural network
DATADIR = r'..\labels\Labelled Dataset - Fig 51'
for DATADIR in DATADIR_LIST:
    for directoryfile in os.listdir(DATADIR):
        if(directoryfile in ignore_files or not os.path.isdir(os.path.join(DATADIR,directoryfile))):
            continue
        CATEGORIES.add(directoryfile) #Append all the character's name as label in 'CATEGORIES'
CATEGORIES = list(CATEGORIES)
print(len(CATEGORIES))
```

```python
CATEGORIES
```

```python
training_data = []

def create_training_data():
    for DATADIR in DATADIR_LIST:
        for category in CATEGORIES :
            path = os.path.join(DATADIR, category)
            # check if the path exist
            if not os.path.exists(path):
                # print(f'Path {path} does not exist')
                
                continue
            class_num = CATEGORIES.index(category) # Charcater's name label as class number
              
            for img in os.listdir(path):
                try :
                    img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE) # Converting image into grayscale
                    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) #Resizing all images into same size (50,50)
                    training_data.append([new_array, class_num]) # append all images with their respective class name
                except Exception as e:
                    print(path)
                    pass

create_training_data()

```

```python
len(training_data)
```

```python
#Storing features in X and labels in Y using numpy

random.shuffle(training_data)
X = [] #features
y = [] #labels

for features, label in training_data:
    X.append(features)
    y.append(label)


X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
#Saving features and label in pickle files

pickle_out = open("X_new.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y_new.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
# load categories from a file
pickle_in = open("categories.pickle", "rb")
CATEGORIES_NEW = pickle.load(pickle_in)
print(CATEGORIES_NEW)
```

```python
import tensorflow as tf
def prepare(file):
    IMG_SIZE = 50
    img_array = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
model = tf.keras.models.load_model("CNN_new.model") # Loding pre-trained data from local machine

input_folder = r"..\labels\Labelled Dataset - Fig 6"
total = 0
correct = 0
for folder in os.listdir(input_folder):
    # check if the folder is a directory
    if not os.path.isdir(os.path.join(input_folder, folder)):
        continue
    for img in os.listdir(os.path.join(input_folder, folder)):

        image = prepare(os.path.join(input_folder, folder, img))
        prediction = model.predict([image])
        prediction = list(prediction[0])
        c = CATEGORIES_NEW[prediction.index(max(prediction))]
        total += 1
        if c == folder:
            correct += 1
```

```python
print("Total: ", total)
print("Correct: ", correct)
print("Accuracy: ", correct/total)
```
